```python
import matplotlib.pyplot as plt
```

```python
cont = ['Anger', 'AverageReinforcement', 'Disgust',
        'Fear', 'Happiness', 'Sadness', 'Status', 'Surprise', 'TotalReinforcement']
```

```python
for prop in cont:
    img1 = plt.imread('Plots/' + prop + '_0.5.png')
    img2 = plt.imread('Plots/' + prop + '_0.75.png')

    fig = plt.figure(figsize = (20, 10))
    ax1 = fig.add_subplot(1,2,1)
    ax1.imshow(img1)
    ax2 = fig.add_subplot(1,2,2)
    ax2.imshow(img2)
    ax1.get_xaxis().set_visible(False)
    ax1.axes.get_yaxis().set_visible(False)
    ax2.get_xaxis().set_visible(False)
    ax2.axes.get_yaxis().set_visible(False)
    fig.tight_layout()
    plt.savefig(prop + '_Comparison.png', dpi=300, bbox_inches = 'tight',
        pad_inches = .5)
    print(prop + ' Done!')
```
